```python
import os
import json
import cv2
import matplotlib.pyplot as plt

# 
image = cv2.imread("/home/u2021213565/jupyterlab/alibaba_qwen/jpg_init_test/2024-07-16T08_30_00Z_below.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# json
with open("/home/u2021213565/jupyterlab/alibaba_qwen/jpg_init_test/2024-07-16T08_30_00Z_below.json", 'r') as f:
    json_data = json.load(f)

# 
for manhole in json_data['manhole']:
    coordinates = manhole['coordinates']
    x1 = coordinates['x1']
    y1 = coordinates['y1']
    x2 = coordinates['x2']
    y2 = coordinates['y2']

    # 
    cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Matplotlib
plt.imshow(image_rgb)
plt.title('Image with Bounding Box')
plt.axis('off')
plt.show()
```
